# Librerías y df

```python
import pandas as pd
```

Leemos el dataframe sacado de https://votaciones.diputados.gob.ar/

```python
df = pd.read_csv("Estadística - 2023.csv")
df
```

# Limpieza de df

Analizamos sus columnas

```python
df.dtypes
```

Reemplazamos los "-" por "0" y cambiamos a "int" las columnas numéricas

```python
df["AFIRM."] = df["AFIRM."].str.replace("-", "0").astype(int)
df["NEG."] = df["NEG."].str.replace("-", "0").astype(int)
df["ABST."] = df["ABST."].str.replace("-", "0").astype(int)
df["AUS."] = df["AUS."].str.replace("-", "0").astype(int)
df["PRESI."] = df["PRESI."].str.replace("-", "0").astype(int)
```

```python
df.dtypes
```

Analizamos la columna "BLOQUE"  y notamos que hay partidos similares.
Reemplazamos sus nombres para poder unirlos.

```python
df["BLOQUE"].value_counts()
```

```python
df["BLOQUE"] = df["BLOQUE"].str.replace("Ucr - Union Civica Radical", "Ucr").str.replace("Frente De Todos", "Union Por La Patria").str.replace("Avanza Libertad" , "La Libertad Avanza")
df["BLOQUE"].value_counts()
```

Armamos una nueva columna "C. SEC" para saber la Cantidad de Sesiones a las que debieron participar durante el año.

```python
df["C. SEC"] = df["AFIRM."]	+ df["NEG."]	+ df ["ABST."]	+ df["AUS."]
df["C. SEC"].value_counts()
```

```python
df.head(3)
```
