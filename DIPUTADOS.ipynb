# Limpieza de df

Analizamos sus columnas

```python
df.dtypes
```

```python
def Replace_and_convert(df, cols):
  for col in cols:
    df[col] = df[col].str.replace("-", "0").astype(int)

cols = ['AFIRM.', 'NEG.', 'ABST.', 'AUS.', 'PRESI.']
Replace_and_convert(df, cols)
```

```python
df.dtypes
```

Reemplazamos los "-" por "0" y cambiamos a "int" las columnas numéricas

```python
df.head(10)
```

Armamos una nueva columna "CANT_SES" para saber la Cantidad de Sesiones a las que debieron participar durante el año.

```python
df["CANT_SES"] = df["AFIRM."]	+ df["NEG."]	+ df ["ABST."]	+ df["AUS."]
df["CANT_SES"].value_counts()
```
